```python
import seaborn as sns
import matplotlib.pyplot as plt

# Function to create scatter plots
def create_scatter_plot(y_values, comp_1_values, comp_2_values, title, filename):
    dfnew = pd.DataFrame({'y': y_values, 'comp-1': comp_1_values, 'comp-2': comp_2_values})

    # Set color palette for color-blind accessibility
    color_palette = sns.color_palette("colorblind", n_colors=len(dfnew['y'].unique()))

    # Adjust the size of the plot
    plt.figure(figsize=(10, 6))

    # Create scatter plot
    sns.scatterplot(sizes=(400, 400), x="comp-1", y="comp-2", hue=dfnew['y'].tolist(),
                    data=dfnew, palette=color_palette, legend='full').set(title=title)

    # Order legend alphabetically
    handles, labels = plt.gca().get_legend_handles_labels()
    order = np.argsort(labels)
    plt.legend(handles=[handles[idx] for idx in order], labels=[labels[idx] for idx in order],
               bbox_to_anchor=(1.05, 0.5), loc='center left', borderaxespad=0., title='CPC Section', markerscale=1)

    # Adjust axis labels and ticks
    plt.xlabel(' ')
    plt.ylabel(' ')
    plt.xticks([])
    plt.yticks([])

    # Save the plot
    plt.savefig(filename, bbox_inches='tight', dpi=300)

    # Show the plot (if needed)
    plt.show()

# Plot for A, Y, and B sections
create_scatter_plot(y_AYB, z[:, 0], z[:, 1], "T-SNE projection of the patent topic vectors for sections A, B, and Y",
                    "C:/Users/bagheri/Desktop/results/new result/fig4_AYB final .png")

# Plot for all sections except A, Y, and B
create_scatter_plot(y_others, z[:, 0], z[:, 1],
                    "T-SNE Projection of Patent Topic Vectors for All Sections Except A, Y, and B",
                    "C:/Users/bagheri/Desktop/results/new result/fig4_no_final.png")
```

```python
def accuracy_method(real_cpc,predict_cpc):
    l=[]
    for i in range(len(predict_cpc)):
        l.append(predict_cpc[i][1])
        
        
    count=len(set(real_cpc)&set(l))
   
    accuracy=(count/(min(len(real_cpc),len(predict_cpc))))*100   
            
    return accuracy
```

```python
def find_cpc_group(cpc_group,level):  ## this code find cpc levels
    classes=cpc_group.split(',')
    #print(classes)

    if level==4:
        return classes
    elif level==1:
        return [ C[0] for C in classes ]
    elif level==3:
        return [C[:4] for C in classes]
    elif level ==2:
        return [C[:3] for C in classes]
    else:
        print("Wrong level!")
```

```python
def find_cpc_group(number_patent,level):  ## this code find cpc levels
    classes=real_cpcnew['cpc_group'][number_patent].split(',')
    #print(classes)

    if level==4:
        a=(classes)
        nn=set(a)
        return list(nn)
     
    elif level==1:
        a=[ C[0] for C in classes ]
        nn=set(a)
        return list(nn)
         
    elif level==3:
        
        a=[C[:4] for C in classes]
        nn=set(a)
        return list(nn)
     
    elif level ==2:
        a= [C[:3] for C in classes]
        nn=set(a)
        return list(nn)
    else:
        print("Wrong level!")
    
```
